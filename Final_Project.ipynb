REGIONAL DISTRIBUTION ANALYSIS USING VALUE COUNTS.

```python
Data["Region"].value_counts().to_frame().style.background_gradient(cmap='twilight')
```

- SHOWS THE DISTRIBUTION OF THE HAPPINESS SCORE VALUES USING BARS.
- KDE(KERNEL DENSITY ESTIMATE) :- KDE LINE SHOWS THE TREND/PATTERN OF THOSE SCORES.

```python
plt.figure(figsize=(12, 6)) 
sns.histplot(Data["Happiness Score"], kde=True)
plt.title("Distribution of Happiness Score", fontsize=16)
plt.xlabel("Happiness Score", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```

COMPARISON OF COUNTRY COUNTS ACROSS DIFFERENT REGIONS.

```python
Data["Region"].value_counts().plot(kind="barh", figsize=(12,6),color = 'teal')
plt.title("Number of Countries per Region")
plt.xlabel("Number of Countries")
plt.ylabel("Region")
plt.show()
```

RELATIONSHIP BETWEEN HEALTH (LIFE EXPECTANCY) AND HAPPINESS SCORE.

```python
plt.figure(figsize=(14, 6))
sns.scatterplot(x="Health (Life Expectancy)",y="Happiness Score",data=Data)
plt.title("Health vs Happiness Score", fontsize=16)
plt.xlabel("Health (Life Expectancy)", fontsize=14)
plt.ylabel("Happiness Score", fontsize=14)
plt.show()
```

CORRELATION HEATMAP OF WORLD HAPPINESS DATASET FEATURE.

```python
corr = Data.corr(numeric_only=True) 

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Happiness Dataset Features", fontsize=16)
plt.show()
```

BOX PLOT ANALYSIS OF HAPPINESS SCORES ACROSS DIFFERENT REGIONS.

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x="Region",y="Happiness Score",data=Data)
plt.xticks(rotation=45)
plt.title("Happiness Score by Region", fontsize=16)
plt.xlabel("Region", fontsize=14)
plt.ylabel("Happiness Score", fontsize=14)
plt.show()
```

BOX PLOT OF HEALTH (LIFE EXPECTANCY) TO IDENTIFY OUTLIERS.

```python
plt.figure(figsize=(14,5))
sns.boxplot(x=Data["Health (Life Expectancy)"], color="skyblue")
plt.xlabel("Health (Life Expectancy)", fontsize=12)
plt.title("Outlier Detection in Health (Life Expectancy)", fontsize=14)
plt.show()
```
